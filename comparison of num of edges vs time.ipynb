```python
import os
import plotly.plotly as py
import plotly.graph_objs as go

my_file=open('/home/neil/Documents/IIITH/sem1/APS/CS3000-Project/compare_dsu.txt','r')
read_data = my_file.readlines()
#print(read_data)
my_file.close()
list1 = []
list2 = []
for lines in read_data:
    l = lines
    x,y = l.split(' ')
    list1.append(float(x))
    list2.append(float(y))
    #print(x)
    #print(y)
dsu = go.Scatter(
    x=list2,
    y=list1, 
    mode='lines+markers',
    name='dsu'
)
my_file=open('/home/neil/Documents/IIITH/sem1/APS/CS3000-Project/compare_veb.txt','r')
read_data = my_file.readlines()
#print(read_data)
my_file.close()
list1 = []
list2 = []
for lines in read_data:
    l = lines
    x,y = l.split(' ')
    list1.append(float(x))
    list2.append(float(y))
    #print(x)
    #print(y)
veb = go.Scatter(
    x=list2,
    y=list1,
    mode='lines+markers',
    name = 'veb'
)
my_file=open('/home/neil/Documents/IIITH/sem1/APS/CS3000-Project/compare_fib.txt','r')
read_data = my_file.readlines()
#print(read_data)
my_file.close()
list1 = []
list2 = []
for lines in read_data:
    l = lines
    x,y = l.split(' ')
    list1.append(float(x))
    list2.append(float(y))
    #print(x)
    #print(y)
fib = go.Scatter(
    x=list2,
    y=list1,
    mode='lines+markers',
    name='fib'
)
```

```python
data = [dsu, veb, fib]
layout = go.Layout(
    title='edge weights inserted in random order',
    xaxis=dict(
        title='edges'
    ),
    yaxis=dict(
        title='time in sec'
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='random_data')
```
